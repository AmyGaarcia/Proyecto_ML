#### Reporte de variables

```python
def data_report(ob_train):
    cols = pd.DataFrame(ob_train.columns.values, columns=["COL_N"])
    types = pd.DataFrame(ob_train.dtypes.values, columns=["DATA_TYPE"])

    percent_missing = round(ob_train.isnull().sum() * 100 / len(ob_train), 2)
    percent_missing_df = pd.DataFrame(percent_missing.values, columns=["MISSINGS (%)"])

    unicos = pd.DataFrame(ob_train.nunique().values, columns=["UNIQUE_VALUES"])
    
    percent_cardin = round(unicos['UNIQUE_VALUES']*100/len(ob_train), 2)
    percent_cardin_df = pd.DataFrame(percent_cardin.values, columns=["CARDIN (%)"])

    concatenado = pd.concat([cols, types, percent_missing_df, unicos, percent_cardin_df], axis=1, sort=False)
    concatenado.set_index('COL_N', drop=True, inplace=True)


    return concatenado.T
        
data_report(ob_train)
```

### REGRESION LOGISTICA

#### train y test

```python
x= ob_train.drop("NObeyesdad_encoded",axis=1)
y= ob_train["NObeyesdad_encoded"].values
x_train, x_test,y_train, y_test =train_test_split(x,y,test_size=0.2, random_state=42)
```

```python
encoder= LabelEncoder()

for columna in x_train.columns:
    if x_train[columna].dtype== "object":
        x_train[columna] =encoder.fit_transform(x_train[columna])
        x_test[columna]= encoder.transform(x_test[columna])
```

#### Escalados

```python
esc= StandardScaler()
min_esc=MinMaxScaler()

train_esc=esc.fit_transform(x_train)
train_min_esc = min_esc.fit_transform(x_train)

test_esc= esc.transform(x_test)
test_min_esc=min_esc.transform(x_test)
```

```python
lasso= LogisticRegression(penalty="l1",solver="liblinear",C=1.0)
lasso.fit(x_train, y_train)
```

#### PCA

```python
pipeline= Pipeline([
    ("pca",PCA(n_components=2)),
    ("clf",SVC())
])
```

#### Modelo

```python
scores= cross_val_score(pipeline,x_train,y_train, cv=5)
print("Cross Validation Scores:",scores)
print("Mean Accuracy:", scores.mean())
```

```python
pipeline.fit(x_train, y_train)
y_pred= pipeline.predict(x_test)
test_balanced_accuracy= balanced_accuracy_score(y_test,y_pred)
print("Test Balanced Accuracy:",test_balanced_accuracy)
```

```python
models= {
    "Linear Regression":LinearRegression(),
    "Random Forest Regressor":RandomForestRegressor(),
    "SVR":SVR(),
    "Logistic Regression":LogisticRegression(),
    "KNN":KNeighborsClassifier(),
    "Decision Tree": DecisionTreeClassifier(),
    "XGBoost":XGBClassifier(),
    "LGBMClassifier":LGBMClassifier(),
    "GradientBoostingClassifier": GradientBoostingClassifier(), 
    "AdaboostClassifier":AdaBoostClassifier(),
    "Catboost": CatBoostClassifier(verbose=False)}

result=[]

for name, model in models.items():
    pipeline.set_params(clf=model)
    pipeline.fit(x_train,y_train)
    train_score= pipeline.score(x_train, y_train)
    test_score= pipeline.score(x_test,y_test)
    result.append({"Modelo":name,"Entrenamiento":train_score,"Puntuaci√≥n":test_score})

ob_train= pd.DataFrame(result)
```

```python
ob_train.sort_values("Puntuaci√≥n",ascending=False)
```

```python
param_grid={
    "KNN":{
        "model":models["KNN"],
        "param_grid":{
        "n_neighbors": [3,5,10],
        "weights":["uniform","distance"],
        "metric":["euclidean","manhattan"]}},
    "Decision Tree":{
        "model":models["Decision Tree"],
        "param_grid":{
        "max_depth":[None, 5,10],
        "min_samples_split": [2, 5,10],
        "min_samples_leaf": [1,2,4]}},
    "Random Forest": {
        "model":models["Random Forest Regressor"],
        "param_grid": {
        "n_estimators": [50,100,200],
        "max_depth": [None, 10, 20],
        "min_samples_split": [2, 5,10],
        "min_samples_leaf": [1,2, 4]}}
}


for model_name, params in param_grid.items():
    model=params["model"]
    param_grid=params["param_grid"]
    
    grid_search= GridSearchCV(model,param_grid, cv=5,scoring="accuracy", n_jobs=-1)
    grid_search.fit(x_train, y_train)
    
    print(f"Mejores hiperpar√°metros para {model_name}:{grid_search.best_params_}")
```

```python
knn= KNeighborsClassifier()
knn.fit(x_train,y_train)

decision_tree= DecisionTreeClassifier()
decision_tree.fit(x_train,y_train)

random_forest= RandomForestClassifier()
random_forest.fit(x_train,y_train)

knn_score =knn.score(x_test,y_test)
dt_score= decision_tree.score(x_test, y_test)
randomf_score= random_forest.score(x_test, y_test)

print("KNN Accuracy:",knn_score)
print("Decision Tree Accuracy:",dt_score)
print("Random Forest Accuracy:",randomf_score)
```

```python
final_decisiontree= DecisionTreeClassifier(max_depth=None, min_samples_leaf=1,min_samples_split=2)

final_decisiontree.fit(x_train,y_train)

dtree_score= final_decisiontree.score(x_test,y_test)
print("Final Decision Tree Accuracy:",dtree_score)
```

```python
y_pred= final_decisiontree.predict(x_test)

accuracy= accuracy_score(y_test, y_pred)
precision= precision_score(y_test, y_pred, average="weighted")
recall=recall_score(y_test, y_pred, average="weighted")
f1= f1_score(y_test, y_pred, average="weighted")

print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1-score:", f1)
```
